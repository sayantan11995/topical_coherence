```python

import os 
import xmltodict
import collections
from dateparser.search import search_dates
import re
import dateparser
import os.path
import fnmatch
import datefinder
from date_extractor import extract_dates
from datetime import datetime
import json
```

```python
with open('config.json') as handle:
    config = json.load(handle)
book = config['book_name']
Parts = config['number_of_parts']
dir_path = './%s/part'% book
```

```python
with open('json_files/%s/year_mode_dict_cassigned.json' %book, 'r', encoding='utf-8') as handle:
    year_mode_dict = json.load(handle)
    
with open('json_files/%s/nltk_chapter_entities_map.json' % book, 'r', encoding='utf-8') as handle:
    nltk_chapter_entity_map = json.load(handle)
    
with open('json_files/%s/date_person_list.json' %book) as handle:
    date_person_list= json.load(handle)
    
with open('json_files/%s/final_person_list.json' % book) as handle:
    people_list_map = json.load(handle)
```

```python
my_dict = {}
# print(people_list_map)
for part in range(1,Parts+1):
    my_dict[str(part)] = {}
    no_of_chapters = len(fnmatch.filter(os.listdir(dir_path + str(part) + '/'), '*.txt'))
#     for i in range(1, no_of_chapters):
    for chapter, entities in nltk_chapter_entity_map[str(part)].items():
        my_dict[str(part)][str(chapter)] = {}
        with open(dir_path + str(part) + '/chapter'+ str(chapter) + '.txt', 'r', encoding='utf-8') as content_file:
                content = content_file.read()
        filtered_entity = people_list_map[str(part)]
        
        for entity in filtered_entity:
            if content.count(entity) != 0:
                my_dict[str(part)][str(chapter)][entity] = content.count(entity) 
#         print(chapter, filtered_entity)
```

```python
year_entity_mention_dict = {}
for part in range(1,Parts+1):
#     year_entity_mention_dict[str(part)] = {}
    no_of_chapters = len(fnmatch.filter(os.listdir(dir_path + str(part) + '/'), '*.txt'))
    for chapter, year in year_mode_dict[str(part)].items():
        
        if year not in year_entity_mention_dict:
            year_entity_mention_dict[year] = my_dict[str(part)][chapter]
        else:
            for entity, mention in my_dict[str(part)][chapter].items():
                try:
                    year_entity_mention_dict[year][entity] += mention
                except:
                    year_entity_mention_dict[year][entity] = mention
```

```python
with open('json_files/%s/date_person_list_with_mention.json' %book, 'w') as handle:
    json.dump(year_entity_mention_dict, handle, indent=4)
```

```python
def get_graph_data(book, year_range):
    
#     with open('app/static/json_files/' + book + '/temporal_year_config.json' , encoding="utf-8") as data_file:
#             temporal_config = json.load(data_file)['temporal_year_list']


    # using category10 color pallete
    color_pallete =  ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

#     people_year_range_list = [ str(item[0]) + '-' + str(item[1]) for item in temporal_config]

    start_year = year_range.split('-')[0]
    end_year = year_range.split('-')[1]

    graph_file_path = "Results/" + book + "/graph_person_startyear_" + start_year + '_endyear_' + end_year + ".xml"
   
    with open(graph_file_path, 'r', encoding="utf8") as xml_file:
        data_dict = xmltodict.parse(xml_file.read())
    xml_file.close()

    
    json_data = json.dumps(data_dict)
    chart_data = []
    node_color_list = []
#     node_community_dict = {}
    for items in data_dict['graphml']['graph']['node']:
        dic = {}
        node = items['@id']
        dic['from'] = node
#         dic['color'] = color_pallete[int(items['data']['#text'])]
        node_color_list.append(dic)
#         node_community_dict[dic['from']] = int(items['data']['#text'])
        

    for edge in data_dict['graphml']['graph']['edge']:
        dic = {}
        source = edge["@source"]
        target = edge["@target"]
        dic['from'] = source
        dic['to'] = target
        
        dic['value'] = 1
        chart_data.append(dic)
        
    node_color_list = sorted(node_color_list, key = lambda i: i['color'])
    chart_data = node_color_list + chart_data

    title = "Mention network of people (" + start_year + "-" + end_year +  ")" 

    return chart_data
```
