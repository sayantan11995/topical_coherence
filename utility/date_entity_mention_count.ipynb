```python
import json, os
```

```python
with open('config.json') as handle:
    config = json.load(handle)
book = config['book_name']
Parts = config['number_of_parts']
dir_path = './%s/part'% book
```

```python
"""
"entity": {

        "part" : {
            "chapterX": [mentions]
        }
        "5": {
            "7": [
                "I was the guest of the late <entity>."
            ]
        }, 
        "4": {}
    }
"""
with open('json_files/%s/mentions.json' %book) as handle:
    mention_list = json.load(handle)

with open('json_files/%s/year_mode_dict_cassigned.json' % book) as handle:
    chapter_year_dict = json.load(handle)
    
with open('json_files/%s/date_place_list.json' % book) as handle:
    date_place_list = json.load(handle)

with open('json_files/%s/date_person_list.json' % book) as handle:
    date_person_list= json.load(handle)
```

```python
date_entity_mention_count = {}
for part, chapters in chapter_year_dict.items():
    for chapter, year in chapters.items():
        date_entity_mention_count[chapter_year_dict[part][chapter]] = {}
#     date_entity_mention_count[chapter_year_dict[part][chapter]] = {}
for entity, part_mention in mention_list.items():
    for part, chapter_mention in part_mention.items():
        for chapter, mention_item_list in chapter_mention.items():
            print(chapter_year_dict[part][chapter], entity, part, chapter, len(mention_item_list))

            try:
                date_entity_mention_count[chapter_year_dict[part][chapter]][entity] += len(mention_item_list)
            except:
                date_entity_mention_count[chapter_year_dict[part][chapter]][entity] = len(mention_item_list)
            
```

```python
date_place_with_mentions = {}
date_person_with_mentions = {}
# for year, entity_list in date_person_list.items():
#     date_person_with_mentions[year] = {}
#     for entity in entity_list:
#         date_person_with_mentions[year][entity] = 0
for year, entity_mentions in date_entity_mention_count.items():
    date_place_with_mentions[year] = {}
    date_person_with_mentions[year] = {}
    for entity in date_person_list[year]:
        if entity in entity_mentions.keys():
            date_person_with_mentions[year][entity] = date_entity_mention_count[year][entity]
    
    for entity in date_place_list[year]:
        if entity in entity_mentions.keys():
            date_place_with_mentions[year][entity] = date_entity_mention_count[year][entity]
```

```python
with open('json_files/%s/date_person_with_mentions.json' %book, 'w') as handle:
    json.dump(date_person_with_mentions,handle,indent = 4)
    
with open('json_files/%s/date_place_with_mentions.json' %book, 'w') as handle:
    json.dump(date_place_with_mentions,handle,indent = 4)
```
